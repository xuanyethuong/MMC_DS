```python
'''
Task 1:
1.1. Tạo một chương trình Python mới có tên “lastname_firstname_grade_the_exams.py.”
1.2. Viết hàm cho phép người dùng nhập tên của một tệp và truy cập đọc.
1.3. Nếu tệp t\ồn tại, bạn có thể in ra một thông báo xác nhận. Nếu tệp không tồn tại, hiển thị không thể tìm thấy tệp và yêu cầu nhập lại.
1.4. Sử dụng try/except để thực hiện việc này (đừng chỉ sử dụng một loạt câu lệnh “if”):
Task 2:
Tiếp theo, bạn sẽ cần phân tích dữ liệu có trong tệp bạn vừa mở để đảm bảo rằng nó ở đúng định dạng. Mỗi tệp dữ liệu chứa một loạt câu trả lời của học sinh ở định dạng sau:
N12345678,B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D
hoặc
N12345678,B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D,A,B,C,D,E
Nhiệm vụ của bạn cho phần này của chương trình là thực hiện như sau:
2.1. Báo cáo tổng số dòng dữ liệu được lưu trữ trong tệp.
2.2. Báo cáo tổng số dòng dữ liệu không hợp lệ trong tệp.
Một dòng hợp lệ chứa danh sách 26 giá trị được phân tách bằng dấu phẩy N# cho một học sinh là mục đầu tiên trên dòng. Nó phải chứa ký tự “N” theo sau là 8 ký tự số. 2.3. Nếu một dòng dữ liệu không hợp lệ, bạn nên báo cáo cho người dùng bằng cách in ra một thông báo lỗi.
Task 3: viết chương trình để chấm điểm các bài thi cho một phần nhất định. Kỳ thi gồm 25 câu hỏi, trắc nghiệm. Đây là một chuỗi đại diện cho các câu trả lời:
answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"
Chương trình sẽ sử dụng những câu trả lời này để tính điểm cho mỗi dòng dữ liệu hợp lệ. Điểm có thể được tính như sau:
+4 điểm cho mỗi câu trả lời đúng 0 điểm cho mỗi câu trả lời bị bỏ qua -1 điểm cho mỗi câu trả lời sai Chúng ta cũng cần thống kê các yêu cầu sau:
Đếm số lượng học sinh đạt điểm cao (>80).
Điểm trung bình.
Điểm cao nhất.
Điểm thấp nhất.
Miền giá trị của điểm (cao nhất trừ thấp nhất).
Giá trị trung vị (Sắp xếp các điểm theo thứ tự tăng dần. Nếu # học sinh là số lẻ, bạn có thể lấy giá trị nằm ở giữa của tất cả các điểm (tức là [0, 50, 100] — trung vị là 50). Nếu # học sinh là chẵn bạn có thể tính giá trị trung vị bằng cách lấy giá trị trung bình của hai giá trị giữa (tức là [0, 50, 60, 100] — giá trị trung vị là 55)).
Trả về các câu hỏi bị học sinh bỏ qua nhiều nhất theo thứ tự: số thứ tự câu hỏi - số lượng học sinh bỏ qua - tỉ lệ bị bỏ qua (nếu có cùng số lượng cho nhiều câu hỏi bị bỏ thì phải liệt kê ra đầy đủ).
Trả về các câu hỏi bị học sinh sai qua nhiều nhất theo thứ tự: số thứ tự câu hỏi - số lượng học sinh trả lời sai - tỉ lệ bị sai (nếu có cùng số lượng cho nhiều câu hỏi bị sai thì phải liệt kê ra đầy đủ).
Lưu ý: các giá trị số thực làm tròn 3 chữ số thập phân
Task 4
Cuối cùng, yêu cầu chương trình là tạo một tệp “kết quả” chứa các kết quả chi tiết cho từng học sinh trong lớp. 
Mỗi dòng của tệp này phải chứa số ID của học sinh, dấu phẩy và sau đó là điểm của họ. Bạn nên đặt tên tệp này dựa trên tên tệp gốc được cung cấp 
— ví dụ: nếu người dùng chọn “class1.txt”, bạn nên lưu trữ kết quả trong tệp có tên “class1_grades.txt”
'''
import os
import numpy as np
def read_file_contents():
    while True:
        # Yêu cầu người dùng nhập tên tệp
        global file_name
        file_name = input("Nhập tên của tệp bạn muốn đọc: ")

        try:
            # Kiểm tra xem tệp có tồn tại không
            if os.path.exists(file_name):
                print(f"Tệp '{file_name}' tồn tại.")
                # Mở tệp để đọc
                with open(file_name, 'r') as file:
                    # Đọc nội dung của tệp
                    contents = file.read()
                    print("Nội dung của tệp là:")
                    print(contents)
                break  # Thoát khỏi vòng lặp khi đã thành công
            else:
                print(f"Không thể tìm thấy tệp '{file_name}'. Vui lòng thử lại.")
        except Exception as e:
            print(f"Có lỗi xảy ra khi đọc tệp: {e}")
def read_line(file_name):
    with open(file_name, 'r') as file:
    # Đọc tất cả các dòng trong tệp và lưu vào danh sách lines
        lines = file.readlines()
    # Đếm số lượng dòng bằng cách đếm số phần tử trong danh sách
        total_lines = len(lines)
        print("Tổng số dòng dữ liệu trong tệp là:", total_lines)
def is_valid_line(line):
    # Kiểm tra xem dòng có chứa đúng định dạng không
    # Định dạng yêu cầu: 'N#', theo sau là 8 ký tự số, sau đó là 25 giá trị phân tách bằng dấu phẩy
    parts = line.strip().split(',')
    if len(parts) != 26:
        return False
    if not (parts[0].startswith('N') and len(parts[0]) == 9 and parts[0][1:].isdigit()):
        return False
    return True
def read_invalid_line(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()
    invalid_lines_count = 0
    # Kiểm tra từng dòng trong tệp
    for line_number, line in enumerate(lines, start=1):
        if not is_valid_line(line):
            print(f"Dòng {line_number} không hợp lệ: {line}")
            invalid_lines_count += 1
def read_valid_line(file_name):
    student_answers = {}
    student_name = {}
    with open(file_name, 'r') as file:
        lines = file.readlines()
    for line_number, line in enumerate(lines, start=1):
        line = line.strip()
        if is_valid_line(line):
            student_answers[str(line_number)] = str(line[9:])  # Convert to string
            student_name[str(line_number)] = str(line[:9])  # Convert to string
    return student_answers, student_name
    
def score_answers(answer_key, student_answers):
    correct_answers = answer_key.split(',')
    student_answers = student_answers.split(',')
    num_questions = len(correct_answers)
    scores = []
    unanswered_questions = [0] * num_questions
    wrong_answers = [0] * num_questions
    
    for i in range(num_questions):
        if student_answers[i] == '':
            unanswered_questions[i] += 1
        elif student_answers[i] == correct_answers[i]:
            scores.append(4)
        else:
            scores.append(-1)
            wrong_answers[i] += 1
    
    # Đếm số lượng học sinh đạt điểm cao (>80)
    num_students_high_score = sum(score > 80 for score in scores)
    
    # Tính điểm trung bình
    average_score = np.mean(scores)
    
    # Tìm điểm cao nhất và thấp nhất
    max_score = max(scores)
    min_score = min(scores)
    
    # Tính miền giá trị của điểm
    score_range = max_score - min_score
    
    # Tìm giá trị trung vị
    sorted_scores = sorted(scores)
    num_students = len(sorted_scores)
    if num_students % 2 == 0:
        median_score = (sorted_scores[num_students // 2 - 1] + sorted_scores[num_students // 2]) / 2
    else:
        median_score = sorted_scores[num_students // 2]
    
    # Tìm câu hỏi bị bỏ qua nhiều nhất
    max_unanswered_questions = max(unanswered_questions)
    most_skipped_questions = [(i+1, unanswered_questions[i], unanswered_questions[i]/len(student_answers)) 
                              for i, count in enumerate(unanswered_questions) if count == max_unanswered_questions]
    # Tìm câu hỏi bị trả lời sai nhiều nhất
    max_wrong_answers = max(wrong_answers)
    most_wrong_questions = [(i+1, wrong_answers[i], wrong_answers[i]/len(student_answers)) 
                            for i, count in enumerate(wrong_answers) if count == max_wrong_answers]
    
    return {
        "num_students_high_score": num_students_high_score,
        "average_score": round(average_score, 3),
        "max_score": max_score,
        "min_score": min_score,
        "score_range": score_range,
        "median_score": median_score,
        "most_skipped_questions": most_skipped_questions,
        "most_wrong_questions": most_wrong_questions
    }
def write_grades_file(file_name, results):
    # Tạo tên tệp kết quả
    grades_file_name = file_name.replace('.txt', '_grades.txt')
    # Ghi kết quả vào tệp
    with open(grades_file_name, 'w') as file:
        for idx, result in enumerate(results, start=1):
            student_id = f"N000000{idx:02}"  # Generate student ID
            grade = result['average_score']  # Get average score
            file.write(f"{student_id},{grade}\n")
read_file_contents()
read_line(file_name)
read_invalid_line(file_name)
student_answers,student_name = read_valid_line(file_name)
print (student_answers)
print (student_name)
# Chuỗi đáp án của kỳ thi
answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D"
# Tính điểm cho mỗi học sinh và thu thập các thống kê
results = []
for idx, sa in student_answers.items():
    student_result = score_answers(answer_key, sa)
    results.append(student_result)

# In kết quả
for idx, result in enumerate(results, start=1):
    print(f"Student {idx}:")
    for key, value in result.items():
        if key == "most_skipped_questions" or key == "most_wrong_questions":
            print(f"{key}:")
            for item in value:
                print(f"  Question {item[0]} - Skipped/Wrong: {item[1]}, Ratio: {round(item[2], 3)}")
        else:
            print(f"{key}: {value}")
    print("\n")
write_grades_file(file_name, results)
print(f"Tệp kết quả đã được lưu trong '{file_name.replace('.txt', '_grades.txt')}'.")
```
