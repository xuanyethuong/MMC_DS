# 2. Hiểu dữ liệu

```python
# Yêu cầu: Quan sát các thông tin về hàng và cột trong dữ liệu

# Thông tin về hàng
print(df_google_playstore.info())

# Thông tin về cột
print(df_google_playstore.columns)

# Thông tin về cả hàng và cột
print(df_google_playstore.shape)
print(df_google_playstore.describe())
```

```python
# Yêu cầu: Quan sát 3 hàng đầu tiên trong dữ liệu bằng phương thức
df_google_playstore.head(3)
```

```python
# Yêu cầu: Quan sát 7 hàng cuối trong dữ liệu bằng phương thức
df_google_playstore.tail(7)
```

```python
# Yêu cầu: Quan sát tổng quan hình dạng của dữ liệu bằng phương thức
print(df_google_playstore.shape)
```

Vậy câu lệnh trên trả về một tuple với 2 giá trị (a, b). Từ đó chúng ta có thể hiểu dữ liệu đang có a dòng và b cột

```python
# Yêu cầu: Quan sát kiểu dữ liệu của từng cột trong dữ liệu
print(df_google_playstore.dtypes)
```

```python
# Yêu cầu: Quan sát thông tin của dữ liệu như số dữ liệu từng cột, kiểu dữ liệu, kích thước của bộ dữ liệu
print(df_google_playstore.info())
```

```python
# Yêu cầu: Quan sát một số đặc điểm của trường dữ liệu dạng số
print(df_google_playstore.describe())
```

```python
# Yêu cầu: Từ các cột dữ liệu dạng số, hãy quan sát:
# - Giá trị lớn nhất,
# - Giá trị nhỏ nhất
# - Vị trí của giá trị lớn nhất
# - Vị trí của giá trị nhỏ nhất
# - Trung bình
# - Trung vị
# - Giá trị xuất hiện nhiều nhất
# - Tổng của toàn bộ dữ liệu
for col_name in df_google_playstore.select_dtypes(['float', 'int']).columns:
    print(col_name)
    print('Sum:', df_google_playstore[col_name].sum())
    print('Min:', df_google_playstore[col_name].min())
    print('Max:', df_google_playstore[col_name].max())
    print('Index value min:', df_google_playstore[col_name].idxmin())
    print('Index value max:', df_google_playstore[col_name].idxmax())
    print('Mean:', df_google_playstore[col_name].mean())
    print('Median:', df_google_playstore[col_name].median())
    print('Mode:', df_google_playstore[col_name].mode())
    print()
```
